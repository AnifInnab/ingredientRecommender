```python
import pandas as pd
import numpy as np
import random
from sklearn.model_selection import train_test_split
#import keras
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Activation, Embedding, Flatten
from keras.preprocessing.sequence import pad_sequences
import ast
import tensorflow as tf
```

```python
#Read files
recipes_path = 'data/RAW_recipes.csv'
recipes = pd.read_csv(recipes_path)
recipes_ingr = recipes['ingredients']
recipes_ingr = recipes_ingr.apply(ast.literal_eval)


ingr_pkl_path = 'data/ingr_map.pkl'
ingr_pkl = pd.read_pickle(ingr_pkl_path)
#list of unique ingredients
ingr_tokens = list(set(ingr_pkl['replaced']))
```

```python
ingr_tokens
```

```python
#Extract one indices
x_onehot_indices = []
y_onehot_indices = []
for recipe in recipes_ingr:
    try:
        label_pos = random.randint(0, len(recipe) - 1)
        y_onehot_index = ingr_tokens.index(recipe[label_pos])
        
        remaining_ingr = recipe[:label_pos] + recipe[label_pos+1:]
        indices = list(map(lambda x : ingr_tokens.index(x), remaining_ingr))
    except:
        continue
    y_onehot_indices.append(ingr_tokens.index(recipe[label_pos]))
    x_onehot_indices.append(indices)

```

```python
#pad index lists to later input into embedding layer
x_maxlen = max([len(onehot) for onehot in x_onehot_indices])
X = pad_sequences(x_onehot_indices, maxlen=x_maxlen, padding='post')

#create onehot vectors for labels
no_examples = len(y_onehot_indices)
no_classes = len(ingr_tokens)
y = np.empty([no_examples, no_classes])
for i, y_idx_list in enumerate(y_onehot_indices):
    y_onehot = np.zeros(no_classes)
    y_onehot[y_idx_list] = 1.0
    y[i] = y_onehot
```

```python
#Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#Train Neural network
def get_compiled_model():
#   model = tf.keras.Sequential([
#     tf.keras.layers.Embedding(input_dim=no_classes, output_dim=100, input_length=no_classes),
#     #tf.keras.layers.Dense(no_classes, activation='relu'),
#     #tf.keras.layers.Dense(8, activation='relu'),
#     #tf.keras.layers.Dense(4, activation='relu'),
#     tf.keras.layers.Flatten(),
#     tf.keras.layers.Dense(no_classes, activation='sigmoid')
#   ])




#     model.add(Embedding(input_dim=own_embedding_vocab_size, # 10
#                     output_dim=32, 
#                     input_length=maxlen))

#   model.compile(optimizer='adam',
#                 loss='categorical_crossentropy',
#                 metrics=['accuracy'],
#                 lr=0.25)

    #model using embeddings
    model = Sequential([
            Embedding(input_dim=no_classes,
                    output_dim=100, 
                    input_length=x_maxlen),
            Flatten(),
            Dense(no_classes, activation='sigmoid')
        ])
    model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

    return model


model = get_compiled_model()
print('hello')
model.fit(x=X_train, y=y_train, epochs=40, batch_size=256)
```
